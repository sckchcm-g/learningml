## Data Visualization With Seaborn

Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code. In this lesson, we will cover the basics of Seaborn, including creating various types of plots and customizing them.

```python
import seaborn as sns
```

```python
# basic plotting with seaborn
tips = sns.load_dataset('tips')
tips
```

```python
# create a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

```python
## Line Plot

sns.lineplot(x='size', y='total_bill', data=tips)
plt.title('Total Bill vs. Size')
```

```python
## Categorical Plot
# bar plot
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Bar plot of Total Bill by Day')
```

```python
## Box Plot
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot of Total Bill by Day')
```

```python
## Violin Plot
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Violin Plot of Total Bill by Day')
```

```python
## Histograms
sns.histplot(tips['total_bill'], bins=10, kde=True)
plt.title('Histogram of Total Bill with KDE')
```

```python
## kde plot
sns.kdeplot(tips['total_bill'], fill=True)
plt.title('KDE Plot of Total Bill')
```

```python
## PairPlot
sns.pairplot(tips)
```

```python
## heatmap
corr = tips[['total_bill', 'tip', 'size']].corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
import pandas as pd
```

```python
sales_df = pd.read_csv('SsalesData.csv')
sales_df.head(5)
```

```python
# plot units sold by item type
plt.figure(figsize=(10,6))
sns.barplot(x='Item Type', y='Units Sold', data=sales_df, estimator=sum)
plt.title('Units Sold by Item Type')
plt.xlabel('Item Type')
plt.ylabel('Total Units Sold')
plt.show()
```

```python
# plot units sold by item type
plt.figure(figsize=(10,6))
sns.barplot(x='Region', y='Units Sold', data=sales_df, estimator=sum)
plt.title('Units Sold by Region')
plt.xlabel('Region')
plt.ylabel('Total Units Sold')
plt.show()
```
